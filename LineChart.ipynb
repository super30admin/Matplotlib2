```python
#Simple line plots
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4]) 
y = x*2 

plt.plot(x, y) 
plt.show() 
```

```python
#Multiple charts 
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4]) 
y = x*2 

plt.plot(x, y) 
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Any suitable title")
plt.show()


plt.figure()
x1=[2,4,6,8]
y1=[3,5,7,9]
plt.plot(x1,y1,'-.')
plt.xlabel("X1 data")
plt.ylabel("Y1 data")
plt.title("New chart with x1 and y1 data")
plt.show()
```

```python
#Multiple plots on the same axis
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4])
y = x*2

# first plot with X and Y data
plt.plot(x, y)

x1 = [2, 4, 6, 8]
y1 = [3, 5, 7, 9]

# second plot with x1 and y1 data
plt.plot(x1, y1, '-.')

plt.xlabel("X-axis data")
plt.ylabel("Y-axis data")
plt.title('multiple plots')
plt.show()
```

```python
#Fill the area between two plots

import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4])
y = x*2

plt.plot(x, y)

x1 = [2, 4, 6, 8]
y1 = [3, 5, 7, 9]

plt.plot(x, y1, '-.')
plt.xlabel("X-axis data")
plt.ylabel("Y-axis data")
plt.title('multiple plots')

plt.fill_between(x, y, y1, color='green', alpha=0.5)
plt.show()
```
